# Write a Python Program to Check if a Number is Positive, Negative or Zero?


```python
#function definition

def check(a): 
    """Check function to check if the number is positive, negative or zero"""
    
    if a == 0: #comparision                                           
        print("The input value is: ", a)
               
        
    elif a < 0 : #comparision 
        print("The input value is a negative number: ", a)
        
    else  : #comparision 
        print("The input value is a postive number: ",a)
    
        
```

```python
check(0.1)
```

# Write a Python Program to Check if a Number is Odd or Even?

```python
#function definition
def oddeven(a): 
    """function to check if the number is odd or even"""
    
    if a % 2 == 0: #check if the remainder is 0
        print(a, "is an Even Number")
    else:
        print(a, "is an Odd Number")
        
```

# Write a Python Program to Check Leap Year?

A leap year is exactly divisible by 4 except for century years (years ending with 00). The century year is a leap year only if it is perfectly divisible by 400. 

```python
#function definition
def leap(n): 
    """Leap function to check if the input year is LEAP year or not."""
    
    if n % 400 == 0 and n % 100 ==0: #remainder should be zero while dividng with 400 and 100
        print(n, "is a Leap Year")
    
    elif n % 4 == 0 and n % 100 !=0: ##remainder should be zero while dividng with 4 and ramiander should not be zero with 100     
        print(n, "is a Leap Year")   
        
    else:
        print(n, "is not a Leap Year") 
    
        
      
```

```python
leap(2900)
```

# Write a Python Program to Check Prime Number?

A whole number greater than 1 that cannot be exactly divided by any whole number other than itself

```python
#function definition
def prime(c): 
    """Prime is function to find if the input is prime number or not"""
    
    if c == 1: # 1 is not a prime number
        print("Not a prime number")
    elif c >1:
        for i in range(2,c):
            if c % i == 0:                
                print(c,"is not a Prime number")
                break
        else:
            print(c,"is a prime number")
        

            
```

```python
prime(11)
```
