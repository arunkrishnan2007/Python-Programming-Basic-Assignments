# Write a Python Program to Find the Factorial of a Number?


```python
def fact(a): 
    """Fact function to find the factorial""" 
    
    if a<0: #Filtering the negative values
        print("The input value must be greater than zero")
        
    elif type(a)==float or type(a)==complex:
        print("The input value must be integer")
        
    
    else:


        b=1 #b for storing the factorial. inital value assigned as 1

        for i in range(1,a+1): #for loop for iteraing within the range
            b=b*i

        print("The Factorial of",a,"is:",b)

fact(.10)
```

# Write a Python Program to Display the multiplication Table?

```python
#function definition
def multab(c): 
    """multab function is used to create mutliplication table upto 10"""
    
    if type(c)==complex: #filtering the complex number
        print("Enter an integer")
        
    else:    
    
        for i in range(1,11): #range function to iterate
            mul=c*i
            
            print (c,"*",i,"=",mul)

multab(5)
```

# Write a Python Program to Print the Fibonacci sequence?

```python
def fib(d):
    e=0
    f=1
    for i in range(d):
        yield e
        e,f=f,f+e
fib(15)
```

```python
for i in fib(15):
    print(i)
```

# Write a Python Program to Find the Sum of Natural Numbers?

The sum of n natural numbers formula is used to find 1 + 2 + 3 + 4 +..... up to n terms. This is arranged in an arithmetic sequence. Hence we use the formula of the sum of n terms in the arithmetic progression for deriving the formula for the sum of natural numbers.

Sum of Natural Numbers Formula: 

 S= [n(n+1)]/2, where n is the natural number.

```python
#function definition
def sumnum(s): 
    """sumnum function to find the sum of natural numbers"""
    
    sum=(s*(s+1))/2 #Formula for sum of natural numbers
    
    return sum
```

```python
sumnum(100)
```
